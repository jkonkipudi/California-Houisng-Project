10.) Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. 
If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
def total_bedroom_size(value): # Creating a function based on total_bedrooms
  if value <= 10: # for total_bedrooms less than 10 considering "SMALL"
    return "SMALL"
  elif 11<=value <1000: # for total_bedrooms between 11 and 1000 considering "MEDIUM"
    return "MEDIUM"
  elif value>=1000: # for total_bedrooms more than 1000 considering "LARGE"
    return "LARGE"

# Now, we will create a mapping function (total_bedroom_) and use the DataFrame.map() function to create a new column from an existing column
CHP["size"]= CHP["total_bedrooms"].map(total_bedroom_size)
CHP.head()
```
